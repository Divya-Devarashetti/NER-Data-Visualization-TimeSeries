```python
 Named Entity Recognition (NER) using spacy
```

```python
import spacy

# Load English tokenizer, tagger, parser and NER
nlp = spacy.load("en_core_web_sm")
```

```python
#this shows the components from nlp part
nlp.pipe_names
```

```python
# Process whole documents
document = nlp("The World Bank provides low-interest loans, zero to low-interest credits, and grants to developing countries.")

# Find named entities, phrases and concepts
for ent in document.ents:
    print(ent.text, "|", ent.label_, "|", spacy.explain(ent.label_))
```

```python
from spacy import displacy

displacy.render(document, style="ent")
```

```python
nlp.pipe_labels['ner']
```

```python
document = nlp("The World Bank provides low-interest loans, zero to low-interest credits, and grants to developing countries.")

for ent in document.ents:
    print(ent.text, "|", ent.label_)
```

```python
type(document[2:5])
```

```python
from spacy.tokens import Span

s1 = Span(document, 0, 1, label="ORG")
s2 = Span(document, 5, 6, label="ORG")

document.set_ents([s1, s2], default="unmodified")
```

```python
for ent in document.ents:
    print(ent.text, "|", ent.label_)
```

```python
 Data Visualization for Time Series
```

```python
#importing Libraries
import pandas as pd
import quandl
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = quandl.get("WIKI/GOOGL")
```

```python
df.head()
```

```python
# deleting column


df.drop(columns='Ex-Dividend')
```

```python
#Plotting a simple line plot for time series data.

df['Volume'].plot()
```

```python
#let’s plot all other columns using subplot.


df.plot(subplots=True, figsize=(12, 14))
```

```python
# Resampling the time series data based on monthly 'M' frequency
df_month = df.resample("M").mean()
  
# using subplot
fig, ax = plt.subplots(figsize=(15, 8))
  
# plotting bar graph
ax.bar(df_month['2017':].index, 
       df_month.loc['2017':, "Volume"], 
       width=25, align='center')
```

```python
#Differencing is used to make the difference in values of a specified interval.


df.Low.diff(2).plot(figsize=(12, 8))
```

```python
df.High.diff(2).plot(figsize=(12, 5))
```

```python
#The shift function can be used to shift the data before or after the specified time interval.
#In this code, .div() function helps to fill up the missing data values. Actually, div() means division
#Here, we have taken .div(df.Close.shift()), it will divide each value of df to df.Close.shift() to remove null values.


df['Change'] = df.Close.div(df.Close.shift())
df['Change'].plot(figsize=(9, 6), fontsize=15)
```

```python
#We can also take a specific interval of time and plot to have a clearer look. Here we are plotting the data of only 2017.


df['2018']['Change'].plot(figsize=(11, 6))
```
